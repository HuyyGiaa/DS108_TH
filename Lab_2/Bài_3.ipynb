```python
import pandas as pd
import matplotlib.pyplot as plt

link_gsheet = "https://docs.google.com/spreadsheets/d/1XEFg047aSbg3OsEVx9PzmgSxGbCvCidfLiHfsgRS3R0/export?format=csv"
dataset = pd.read_csv(link_gsheet)

dataset['Date Announced'] = pd.to_datetime(dataset['Date Announced'])
```

**3.a) Liệt kê số ca nhiễm theo từng thành phố (Detected City).**

```python
city_counts = dataset['Detected City'].value_counts()
print(city_counts)
```

**3.b) Liệt kê số ca nhiễm theo độ tuổi, vẽ biểu đồ (sử dụng hàm plot).**

```python
# Đếm số ca theo độ tuổi và sắp xếp theo thứ tự nhóm tuổi
age_counts = dataset['Age Bracket'].value_counts().sort_index()
print(age_counts)

plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar', color='orange')
plt.title('Thống kê số ca nhiễm theo độ tuổi', fontsize=14)
plt.xlabel('Nhóm tuổi')
plt.ylabel('Số ca nhiễm')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**3.c) Liệt kê số ca nhiễm tại Hokkaido theo từng ngày. Vẽ biểu đồ.**

```python
# Lọc dữ liệu riêng cho tỉnh Hokkaido
hokkaido_data = dataset[dataset['Detected Prefecture'] == 'Hokkaido']

# Thống kê số ca theo ngày và sắp xếp theo thời gian
hokkaido_by_date = hokkaido_data['Date Announced'].value_counts().sort_index()
print(hokkaido_by_date)

plt.figure(figsize=(12, 6))
hokkaido_by_date.plot(kind='line', marker='o', color='red', linewidth=2)
plt.title('Diễn biến số ca nhiễm tại Hokkaido theo ngày', fontsize=14)
plt.xlabel('Ngày công bố')
plt.ylabel('Số ca nhiễm') 
plt.grid(True, alpha=0.7)
plt.tight_layout()
plt.show()
```
