```python
import pandas as pd
import os
import numpy as np
import urllib.request
import zipfile 
```

# Câu 1

```python
data_url = "https://d396qusza40orc.cloudfront.net/getdata%2Fprojectfiles%2FUCI%20HAR%20Dataset.zip"
data_path = "UCI_HAR_Dataset.zip"
dataset = "UCI HAR Dataset"

# Kiểm tra xem dữ liệu đã tồn tại hay chưa, nếu chưa thì tải về và giải nén
if not os.path.exists(dataset):
    print("Đang tải dữ liệu")
    urllib.request.urlretrieve(data_url, data_path) # Tải dữ liệu từ URL và lưu vào data_path
    print("Tải dữ liệu thành công")
    with zipfile.ZipFile(data_path, 'r') as zip_ref: # Mở file zip và giải nén vào thư mục hiện tại
        zip_ref.extractall(".")
    print("Giải nén thành công")
    os.remove(data_path) # Xóa file zip sau khi giải nén
    print("Đã xóa file zip")
else:
    print("Dữ liệu đã tồn tại")
```

# Câu 3

```python
print("Đang đọc dữ liệu huấn luyện")
# Đọc dữ liệu huấn luyện
x_train = pd.read_csv(os.path.join(dataset, "train", "X_train.txt"), sep="\s+", header=None) 
y_train = pd.read_csv(os.path.join(dataset, "train", "y_train.txt"), header=None, names=["Activity"])
subject_train = pd.read_csv(os.path.join(dataset, "train", "subject_train.txt"), header=None, names=["Subject"])
print("Đọc dữ liệu huấn luyện thành công")

print("Đang đọc dữ liệu kiểm tra")
# Đọc dữ liệu kiểm tra
x_test = pd.read_csv(os.path.join(dataset, "test", "X_test.txt"), sep="\s+", header=None)
y_test = pd.read_csv(os.path.join(dataset, "test", "y_test.txt"), header=None, names=["Activity"])
subject_test = pd.read_csv(os.path.join(dataset, "test", "subject_test.txt"), header=None, names=["Subject"])
print("Đọc dữ liệu kiểm tra thành công")

print("Đang đọc tên các đặc trưng")
# Đọc tên các đặc trưng
features = pd.read_csv(os.path.join(dataset, "features.txt"), sep="\s+", header=None, names=["Index", "Feature"])
print("Đọc tên các đặc trưng thành công")
```

```python
# Gán tên cột cho dữ liệu huấn luyện và kiểm tra
features_names = features["Feature"].values
x_train.columns = features_names
x_test.columns = features_names
```

```python
# Gộp dữ liệu huấn luyện
train_dataset = pd.concat([x_train, y_train, subject_train], axis=1)
print(train_dataset.columns.to_list())
```

```python
# Gộp dữ liệu kiểm tra
test_dataset = pd.concat([x_test, y_test, subject_test], axis=1)
print(test_dataset.columns.to_list())
```

# Câu 4

```python
print("Các cột có chứa mean trong tên:")
# Lọc các cột có chứa "mean" trong tên
key_mean = "mean"
x_train_columns = x_train.columns.to_list()
mean_columns = [col for col in x_train_columns if key_mean in col]
print(mean_columns)
```

```python
print("Các cột chứa std trong tên:")
# Lọc các cột có chứa "std" trong tên
key_std = "std"
std_columns = [col for col in x_train_columns if key_std in col]
print(std_columns)
```

```python
print(x_train.columns.to_list())
```

```python
# Gộp dữ liệu huấn luyện và kiểm tra chỉ với các cột chứa mean và std, cùng với cột "Activity" và "Subject"
std_mean_columns = mean_columns + std_columns + ["Activity", "Subject"]
dataset_std_mean = pd.concat([train_dataset[std_mean_columns], test_dataset[std_mean_columns]], axis=0)
```

# Câu 5

```python
# Tạo một DataFrame mới với tên cột được làm rõ nghĩa hơn
patterns = {
    r'^t': 'Time domain: ',
    r'^f': 'Frequency domain: ',
    r'-': ', ',
    r'mean\(\)': 'mean value',
    r'std\(\)': 'standard deviation value',
    r'-X': ' on X axis',
    r'-Y': ' on Y axis',
    r'-Z': ' on Z axis',
    r'AccJerk': 'acceleration jerk',
    r'Acc': 'acceleration',
    r'GyroJerk': 'angular velocity jerk',
    r'Gyro': 'angular velocity',
    r'Mag': 'magnitude'
}
```

```python
# Đổi tên cột trong dữ liệu huấn luyện
for i, col in enumerate(x_train_columns):
    for pattern, replacement in patterns.items():
        col = pd.Series(col).str.replace(pattern, replacement, regex=True).values[0]
    x_train_columns[i] = col

x_train.columns = x_train_columns
print(x_train.columns.to_list())
```

```python
# Đổi tên cột trong dữ liệu kiểm tra
for i, col in enumerate(x_test.columns):
    for pattern, replacement in patterns.items():
        col = pd.Series(col).str.replace(pattern, replacement, regex=True).values[0]
    x_test.columns.values[i] = col
print(x_test.columns.to_list())
```

```python
tidy_data = dataset_std_mean.groupby(["Subject", "Activity"]).mean().reset_index()
print(tidy_data["Activity"].head())
print(tidy_data.shape)
```

```python
# Khai báo hàm để ánh xạ nhãn hoạt động từ số sang tên
def map_activity_labels(activity):
    activity_labels = {
        1: "WALKING",
        2: "WALKING_UPSTAIRS",
        3: "WALKING_DOWNSTAIRS",   
        4: "SITTING",
        5: "STANDING",
        6: "LAYING"
    }
    return activity_labels[activity]
```

```python
# Ánh xạ nhãn hoạt động trong cột "Activity" của tidy_data
tidy_data["Activity"] = tidy_data["Activity"].apply(map_activity_labels)
print(tidy_data["Activity"].head())
```

```python
# Lưu dữ liệu tidy vào file CSV
tidy_data.to_csv("tidy_data.csv", index=False)
print("Dữ liệu tidy đã được lưu vào tidy_data.csv")
```
