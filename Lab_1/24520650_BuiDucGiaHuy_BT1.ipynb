**g)** Đếm số lượng ca nhiễm mới theo từng tỉnh của Trung Quốc trong tháng 02/2021. 

```python
import pandas as pd

# Sắp xếp dữ liệu theo từng tỉnh và ngày tháng 
coronaChina = coronaChina.sort_values(by=["Province/State", "ObservationDate"]) #Đảm bảo dữ liệu được sắp xếp để lấy ngày sau trừ đi ngày trước

# Tạo thêm một cột để lưu lại số ca nhiễm mới trong ngày của từng tỉnh
coronaChina["New_Cases_In_Day"] = coronaChina.groupby("Province/State")["Confirmed"].diff().fillna(0)

# Lọc dữ liệu cho tháng 2 năm 2021
coronaChina_feb = coronaChina[(coronaChina["ObservationDate"] >= "02/01/2021") & (coronaChina["ObservationDate"] <= "02/28/2021")]

# Tính tổng số ca lây nhiễm mới theo từng tỉnh
New_Cases_per_province = coronaChina_feb.groupby(["Country/Region", "Province/State"])["New_Cases_In_Day"].sum()
print("Số ca lây nhiễm mới theo từng tỉnh của Trung Quốc trong tháng 2 năm 2021:")
print(New_Cases_per_province)
```

**i)** *Có nhận xét gì về số ca nhiễm mới tại Việt Nam giữa tháng 05/2020 và tháng 05/2021. Vẽ biểu đồ đường thể hiện số ca nhiễm mới trong 2 tháng trên. Gợi ý: Dùng hàm plot() trong R.

```python
import matplotlib.pyplot as plt
```

```python
#Đảm bảo cột ngày tháng đúng định dạng (nếu bạn chưa chuyển)
coronaVietNam['ObservationDate'] = pd.to_datetime(coronaVietNam['ObservationDate'])

#Gộp theo ngày và tính số ca nhiễm mới hàng ngày
VietNam_daily = coronaVietNam.groupby('ObservationDate')['Confirmed'].max().reset_index()
VietNam_daily['NewCases'] = VietNam_daily['Confirmed'].diff().fillna(0)

#Lọc dữ liệu cho 2 tháng cần xét
may_2020 = VietNam_daily[(VietNam_daily['ObservationDate'].dt.year == 2020) & (VietNam_daily['ObservationDate'].dt.month == 5)]
may_2021 = VietNam_daily[(VietNam_daily['ObservationDate'].dt.year == 2021) & (VietNam_daily['ObservationDate'].dt.month == 5)]

#Vẽ biểu đồ đường
plt.figure(figsize=(14, 5))

#Đồ thị tháng 05/2020
plt.subplot(1, 2, 1)
plt.plot(may_2020['ObservationDate'], may_2020['NewCases'], marker='o', color='blue')
plt.title('Số ca nhiễm mới tại VN (Tháng 05/2020)')
plt.xlabel('Ngày')
plt.ylabel('Số ca')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

#Đồ thị tháng 05/2021
plt.subplot(1, 2, 2)
plt.plot(may_2021['ObservationDate'], may_2021['NewCases'], marker='o', color='red')
plt.title('Số ca nhiễm mới tại VN (Tháng 05/2021)')
plt.xlabel('Ngày')
plt.ylabel('Số ca')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**k)** * Vẽ biểu đồ về số ca lây nhiễm nhiều nhất của 3 quốc gia: Vietnam, Indonesia và Philippine trong 2 tháng gồm 01 và tháng 02 năm 2021. 

```python
#Lọc tháng 1 và 2 năm 2021
VietNam_jan_feb = VietNam_daily[(VietNam_daily['ObservationDate'] >= '2021-01-01') & (VietNam_daily['ObservationDate'] <= '2021-02-28')]
max_vn = VietNam_jan_feb['NewCases'].max()

#Xử lý cho Indonesia và Philippines từ bảng dữ liệu gốc 
#Đảm bảo df gốc đã chuẩn hóa ngày tháng
data['ObservationDate'] = pd.to_datetime(data['ObservationDate'])

# Lọc 2 quốc gia và khoảng thời gian
mask_country = data['Country/Region'].isin(['Indonesia', 'Philippines'])
mask_date = (data['ObservationDate'] >= '2021-01-01') & (data['ObservationDate'] <= '2021-02-28')
Indo_Phil_df = data[mask_country & mask_date].copy()

# Tính ca nhiễm mới mỗi ngày cho 2 nước này
daily_indo_phil = Indo_Phil_df.groupby(['Country/Region', 'ObservationDate'])['Confirmed'].max().reset_index()
daily_indo_phil['NewCases'] = daily_indo_phil.groupby('Country/Region')['Confirmed'].diff().fillna(0)

# Tìm mức cao nhất
max_others = daily_indo_phil.groupby('Country/Region')['NewCases'].max()

#Gộp dữ liệu 3 nước và vẽ biểu đồ 
countries = ['Vietnam', 'Indonesia', 'Philippines']
peak_cases = [max_vn, max_others['Indonesia'], max_others['Philippines']]

plt.figure(figsize=(8, 6))
bars = plt.bar(countries, peak_cases, color=['#ff9999', '#66b3ff', '#99ff99'])

# Ghi chú giá trị lên đỉnh các cột
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{int(yval)}', ha='center', va='bottom', fontweight='bold')

plt.title('Số ca nhiễm mới cao nhất trong 1 ngày (Tháng 01-02/2021)')
plt.ylabel('Số ca')
plt.show()
```
