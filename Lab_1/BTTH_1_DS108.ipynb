**f)** Thống kê số lượng record theo từng tỉnh của Trung Quốc trong tháng 02/2021.

```python
#Số lượng ca lây nhiễm được xác nhận tại Trung Quốc đại lục (Mainland China) trong tháng 2 năm 2021
coronaChina_feb = coronaChina[(coronaChina["ObservationDate"] >= "02/01/2021") & (coronaChina["ObservationDate"] <= "02/29/2021")]
print(coronaChina_feb)
```

```python
#Thống kế số lượng record theo từng tỉnh của Trung Quốc trong tháng 2 năm 2021
Record_per_province = coronaChina_feb.groupby(["Country/Region", "Province/State"])["SNo"].count()
print("Số lượng record theo từng tỉnh của Trung Quốc trong tháng 2 năm 2021:")
print(Record_per_province)
```

**g)** Đếm số lượng ca nhiễm mới theo từng tỉnh của Trung Quốc trong tháng 02/2021. 

```python
import pandas as pd

# Sắp xếp dữ liệu theo từng tỉnh và ngày tháng 
coronaChina = coronaChina.sort_values(by=["Province/State", "ObservationDate"]) #Đảm bảo dữ liệu được sắp xếp để lấy ngày sau trừ đi ngày trước

# Tạo thêm một cột để lưu lại số ca nhiễm mới trong ngày của từng tỉnh
coronaChina["New_Cases_In_Day"] = coronaChina.groupby("Province/State")["Confirmed"].diff().fillna(0)

# Lọc dữ liệu cho tháng 2 năm 2021
coronaChina_feb = coronaChina[(coronaChina["ObservationDate"] >= "02/01/2021") & (coronaChina["ObservationDate"] <= "02/28/2021")]

# Tính tổng số ca lây nhiễm mới theo từng tỉnh
New_Cases_per_province = coronaChina_feb.groupby(["Country/Region", "Province/State"])["New_Cases_In_Day"].sum()
print("Số ca lây nhiễm mới theo từng tỉnh của Trung Quốc trong tháng 2 năm 2021:")
print(New_Cases_per_province)
```

```python
coronaChina_Deaths = coronaChina[(coronaChina["ObservationDate"] >= "02/01/2021") & (coronaChina["ObservationDate"] <= "02/15/2021")][["Deaths", "ObservationDate"]]
print("Số ca tử vong tại Trung Quốc đại lục trong khoảng thời gian từ 01/02/2021 đến 15/02/2021:")
print(coronaChina_Deaths)
```
