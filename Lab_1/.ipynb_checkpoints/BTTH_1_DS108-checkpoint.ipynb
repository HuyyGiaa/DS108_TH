```python
import pandas as pd
```

```python
data = pd.read_csv("covid_19_data.csv")
```

**Bài tập**

**a)** Code lại các ví dụ trong Phần 3

```python
#Liệt kê số cột trong bảng dữ liệu
num_cols = data.shape[1]
print(num_cols)
```

```python
#Liệt kê số hàng trong bảng dữ liệu
num_row = data.shape[0]
print(num_row)
```

```python
#In ra 10 dòng đầu trong bảng dữ liệu
print(data.head(10))
```

```python
#In ra tên các biến (tên cột) của dữ liệu
print(data.columns)
```

```python
#Tạo biến countryCorona lưu giá trị là các quốc gia có dịch Corona
countryCorona = data["Country/Region"]
print(countryCorona)
```

```python
#Liệt kê số lượng ca lây nhiễm được xác nhận (biến Confirmed) nhiều nhất và lưu vào biến maxConfirmedCases
maxConfirmedCases = max(data["Confirmed"])
print(maxConfirmedCases)
```

```python
#Liệt kê các dữ liệu về covid-19 tại quốc gia Trung Quốc đại lục (Mainland China) và lưu vào biến coronaChina.
coronaChina = data[data["Country/Region"] == 'Mainland China']
print(coronaChina)
```

```python
#Tìm quốc gia (Country.Region) có số ca lây nhiễm nhiều nhất
maxCountryConfirmedCorona = data[data["Confirmed"] == maxConfirmedCases]
print(maxCountryConfirmedCorona["Country/Region"])
```

```python
#Tìm tỉnh (Province.State) có số ca lây nhiễm nhiều nhất
maxStateConfirmedCorona = data[data["Confirmed"] == maxConfirmedCases]
print(maxStateConfirmedCorona["Province/State"])
```

```python
# Lấy dữ liệu theo ngày tháng
Observate_date = data["ObservationDate"]
print(Observate_date)

# Định dạng lại cột dữ liệu ngày tháng
data["ObservationDate"] = pd.to_datetime(data['ObservationDate'])
data["ObservationDate"] = data['ObservationDate'].dt.strftime('%m/%d/%Y')
print(data["ObservationDate"])

#Lấy dữ liệu trong tháng 1/2020: bắt đầu từ 01/01/2020 đến 31/01/2020.
print(data[(data["ObservationDate"] <= "01/31/2020") & (data["ObservationDate"] >= "01/01/2020")])
```

**b)** Tìm dữ liệu về số ca lây nhiễm tại Vietnam (Country.Region == 'Vietnam') và lưu
vào biến coronaVietnam.

```python
coronaVietNam = data[data["Country/Region"] == 'Vietnam']
print(coronaVietNam)
```

**c)** Tìm dữ liệu về số ca lây nhiễm tại Việt Nam trong tháng 02 năm 2021.

```python
#Lấy dữ liệu về số ca lây nhiễm của tháng 2 năm 2021 tại Việt Nam
VietNam_02_2021 = coronaVietNam[(coronaVietNam["ObservationDate"] >= "02/01/2021") & (coronaVietNam["ObservationDate"] <= "02/29/2021")].copy()
print(VietNam_02_2021)
```

**d)** In ra số dữ liệu về ca lây nhiễm nhiều nhất trong khoảng tháng 01 và 02 tại Việt
Nam (Lấy năm 2021).

```python
#Tháng 2
#Tạo thêm một cột để lưu lại số ca nhiễm mới trong ngày
VietNam_02_2021['New_Cases'] = VietNam_02_2021['Confirmed'].diff()
print(VietNam_02_2021[['ObservationDate', 'Confirmed', 'New_Cases']])
VietNam_02_2021['New_Cases'] = VietNam_02_2021['New_Cases'].fillna(0) #Điền 0 vào các ô có giá trị bị NaN

#In ra dữ liệu về ca lây nhiễm nhiều nhất trong tháng 2 năm 2021 tại Việt Nam
max_VietNam_02_2021 = VietNam_02_2021[(VietNam_02_2021["New_Cases"] == max(VietNam_02_2021["New_Cases"]))]
print(max_VietNam_02_2021)

#Tháng 1
#Lấy dữ liệu về số ca nhiễm của tháng 1 năm 2021 ở Việt Nam
VietNam_01_2021 = coronaVietNam[(coronaVietNam["ObservationDate"] >= "01/01/2021") & (coronaVietNam["ObservationDate"] <= "01/31/2021")].copy()
print(VietNam_01_2021)

VietNam_01_2021['New_Cases'] = VietNam_01_2021['Confirmed'].diff()
print(VietNam_01_2021[['ObservationDate', 'Confirmed', 'New_Cases']])
VietNam_01_2021['New_Cases'] = VietNam_01_2021['New_Cases'].fillna(0)

#In ra dữ liệu về ca lây nhiễm nhiều nhất trong tháng 1 năm 2021 tại Việt Nam
max_VietNam_01_2021 = VietNam_01_2021[VietNam_01_2021["New_Cases"] == max(VietNam_01_2021["New_Cases"])]
print(max_VietNam_01_2021)
```
